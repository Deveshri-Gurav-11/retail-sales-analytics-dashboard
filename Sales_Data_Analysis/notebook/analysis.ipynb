```python
df['Month'] = df['Order Date'].dt.month
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales
```

```python
plt.figure()

monthly_sales.plot()

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

Sales increase toward the end of the year, indicating seasonal demand.

```python
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

subcategory_sales
```

```python
plt.figure()

subcategory_sales.plot(kind='bar')

plt.title("Sales by Sub Category")

plt.show()
```

Chairs, Phones, and Storage are among the highest revenue generating sub-categories.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='Discount', y='Profit', data=df)

plt.title("Profit vs Discount")

plt.show()
```

Profit vs Discount Insight

- Most profitable orders occur when no discount is applied.
- Higher discounts tend to reduce profit margins.
- Some orders with large discounts even result in negative profit.
- Businesses should carefully manage discount strategies to avoid profit loss.

```python
top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))

top_cities.plot(kind='bar')

plt.title("Top 10 Cities by Sales")
plt.xlabel("City")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales
```

```python
plt.figure(figsize=(10,6))

monthly_sales.plot()

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.show()
```

Monthly Sales Insight
- Sales fluctuate across months.
- Higher sales are often observed toward the end of the year, indicating seasonal demand.
- Businesses may increase marketing or inventory planning during high-sales months.

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(7,5))

segment_sales.plot(kind='bar')

plt.title("Sales by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Total Sales")

plt.show()
```

The Consumer segment contributes the highest share of sales, followed by Corporate and Home Office customers.
- The West region generates the highest revenue.
- Technology category contributes the highest profit.
- Phones, Chairs, and Storage are top-performing sub-categories.
- Higher discounts reduce profit margins.
- Consumer segment drives most purchases.
- Sales show seasonal trends across months.
